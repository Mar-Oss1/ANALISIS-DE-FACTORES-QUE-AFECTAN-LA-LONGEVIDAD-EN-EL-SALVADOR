```python
import pandas as pd

life = pd.read_csv("/kaggle/input/life-longevity-factors/LiveLongerData.csv")
life = life.drop(['Comment', 'Note', 'Sources', 'Links', 'ID'], axis=1)
life
```

We found, {'suggestive' : 1, 'good' : 2, 'strong' : 3} : `strength of science?` & `strength of science as a number` relation.

```python
df = pd.DataFrame(life)

positive = df[df['Years gained / lost'] > 0]
positive
```

```python
import matplotlib.pyplot as plt

df = pd.DataFrame(positive)
plt.figure(figsize=(12, 6))
plt.bar(df['Factor'], df['Years gained / lost'], color='blue')
plt.xticks(rotation=90)
plt.xlabel('Factor')
plt.ylabel('Years gained / lost')
plt.title('Distribution of Years gained / lost by Factor')
plt.tight_layout()

plt.show()
print(positive['Factor'])
```

```python
negative = life[life['Years gained / lost'] < 0]
negative
```

```python
df = pd.DataFrame(negative)
plt.figure(figsize=(12, 6))
negative_values = df[df['Years gained / lost'] < 0]
plt.bar(negative_values['Factor'], negative_values['Years gained / lost'], color='red', label='Negative')
plt.xticks(rotation=90)
plt.xlabel('Factor')
plt.ylabel('Years gained / lost')
plt.title('Distribution of Years gained / lost by Factor')
plt.legend()
plt.tight_layout()

plt.show()
print(negative_values['Factor'])
```

```python
df = pd.DataFrame(life)

# 
plt.figure(figsize=(12, 6))

# Positive
positive_values = df[df['Years gained / lost'] > 0]
plt.bar(positive_values['Factor'], positive_values['Years gained / lost'], color='blue', label='Positive')

# Negative
negative_values = df[df['Years gained / lost'] < 0]
plt.bar(negative_values['Factor'], negative_values['Years gained / lost'], color='red', label='Negative')

plt.xticks(rotation=90)
plt.xlabel('Factor')
plt.ylabel('Years gained / lost')
plt.title('Distribution of Years gained / lost by Factor')
plt.legend()
plt.tight_layout()

plt.show()
```

El análisis y las visualizaciones muestran que las decisiones de estilo de vida, como no fumar, hacer ejercicio, comer saludablemente, meditar y pasar tiempo con seres queridos, pueden aumentar significativamente la esperanza de vida. Además, evitar el cáncer y las enfermedades cardíacas, tener un buen matrimonio y vivir en el campo también contribuyen a una vida más larga.

Por otro lado, fumar, tener sobrepeso, beber en exceso, estar soltero, tener una perspectiva pesimista y ser un adicto al trabajo están entre los principales factores que disminuyen la esperanza de vida.

Las campañas de salud pública para promover un estilo de vida saludable, el ejercicio regular y la meditación, así como los programas de apoyo a la salud mental y las relaciones saludables, pueden ayudar a aumentar la esperanza de vida. Además, las iniciativas para reducir la contaminación y fomentar la vida en áreas menos pobladas, y las campañas de salud pública para concienciar sobre los riesgos del cáncer y las enfermedades cardíacas y promover chequeos de salud regulares, también pueden contribuir a una vida más larga.
